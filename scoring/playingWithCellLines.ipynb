```python
df2 = pd.read_csv("../drugCombDBdata/synergy-score-data/REGRdrugcombs_scored.csv")
df = df2["Cell line"].str.lower()
df = df.str.replace(" ","")
df = set(df.str.replace("-",""))
del df2
```

```python
gex.columns = gex.columns.str.lower()
gex.columns = gex.columns.str.replace(" ","")
gex.columns = gex.columns.str.replace("-","")
```

```python
shared_genes = np.array(list(set(gex.columns).intersection(df)))
```

```python
gex.set_index("unnamed:0",inplace=True)
```

```python
gex4PCA = gex.loc[:,shared_genes].transpose()
print(gex4PCA.shape)
gex4PCA.head(2)
```

```python
from sklearn.decomposition import PCA

# Why would I be only allowed to make min(features, samples) components? Certainly 24 can then not explain 100% of the variance
pca = PCA(n_components=15)
cell_line_representation = pd.DataFrame(pca.fit_transform(gex4PCA))
```

```python
cell_line_representation.index=gex4PCA.index
```
