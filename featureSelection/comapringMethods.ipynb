```python
# 35 seconds computation
# corrMatrix = data.iloc[:,8:].corr()
# with open('./pickles/correlationMatrix.pkl', 'wb') as outp:
#     pickle.dump(corrMatrix, outp)

with open('./pickles/correlationMatrix.pkl', 'rb') as inp:
    corrMatrix = pickle.load(inp)
```

```python
correlatedFeatures = []
for i in range(len(corrMatrix.columns)):
    for j in range(i):
        if abs(corrMatrix.iloc[i, j]) > 0.0001:
            colname = corrMatrix.columns[i]
            correlatedFeatures.append(colname)
len(correlatedFeatures)
```

```python
smallerData = data.drop(labels=correlatedFeatures,axis=1)
Y = smallerData.iloc[:,6]
X = smallerData.iloc[:,8:]
X.reset_index(inplace=True)
X.drop(["D2 cIds","index"],axis=1, inplace=True)
# expected 533854 rows × 138 columns
X_train,x_test,Y_train,y_test = train_test_split(X,Y,test_size=0.3)

regressor = DecisionTreeRegressor()
regressor.fit(X_train,Y_train)#.drop(["D1 Ipc","D2 Ipc"]

np.mean(abs(regressor.predict(x_test)-y_test))#.drop(["D1 Ipc","D2 Ipc"]
```
